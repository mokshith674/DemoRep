```python
import pandas as pd
df=pd.read_csv('C:\\Users\\GPT BANTWAL\\Downloads\\mtcars.csv')
df
```

```python
import matplotlib.pyplot as plt
range=(10,40)
bins=10
plt.hist(df['mpg'],bins,range,color='red',rwidth=0.8)
plt.title('Histogram')
plt.xlabel('mpg')
plt.ylabel('frequency')
```

```python
plt.scatter(df['wt'],df['mpg'],color='green',marker='*')
plt.title('scatterplot')
plt.xlabel('weight')
plt.ylabel('mpg')
plt.show()
```

```python
temp=df['am'].value_counts()
print(temp)
temp.plot.bar()
plt.title('Barplot')
plt.xlabel('Transmission Types')
plt.ylabel('am')
plt.show()
```

```python
df['mpg'].plot(kind='box')
plt.title('Boxplot')
plt.show()
```
