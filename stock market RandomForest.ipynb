```python
import pandas as pd
df=pd.read_csv("D:\\miniproject\\stocks.csv")
df
```

```python
x=df.drop(['Close','Date','Ticker','Adj Close'],axis=1)
x
```

```python
y=df['Close']
y
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.90,random_state=100)
print(x_train.shape)
print(x_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators=100,random_state=0)
regressor.fit(x_train,y_train)
```

```python
predicted=regressor.predict(x_test)
predicted
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
train_accuracy=regressor.score(x_train,y_train)
print('train_accuracy:',train_accuracy)
R_test_accuracy=regressor.score(x_test,y_test)
print('test_accuracy:',R_test_accuracy)
```
